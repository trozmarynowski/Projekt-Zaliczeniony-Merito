```python

# --- ANALIZA: RASOWY CZY MIESZANIEC? ---
# Sprawdzam, czy zwierzęta rasowe znajdują dom szybciej niż mieszańce.
# Jeśli Breed2 == 0, to znaczy, że zwierzę ma zdefiniowaną tylko jedną rasę.

# Tworzę nową kolumnę tekstową do wykresu
df['Status_Rasy'] = df['Breed2'].apply(lambda x: 'Rasowy / W typie rasy' if x == 0 else 'Mieszaniec')

plt.figure(figsize=(10, 6))

# Rysuję wykres
ax = sns.countplot(x='AdoptionSpeed', hue='Status_Rasy', data=df, palette='autumn')

plt.title('Szybkość adopcji: Rasowe vs Mieszańce', fontsize=15)
plt.xlabel('Szybkość Adopcji (0=Szybko ... 4=Wolno)', fontsize=12)
plt.ylabel('Liczba zwierząt', fontsize=12)
plt.legend(title='Typ')

plt.show()

print(df['Status_Rasy'].value_counts())
```
