```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


# Korzystam z plików umieszczonych na GitHubie


# Poniżej wklejam linki do mojego repozytorium (RAW)
url_train = 'https://raw.githubusercontent.com/trozmarynowski/Projekt-Zaliczeniony-Merito/refs/heads/main/train.csv'
url_breed = 'https://raw.githubusercontent.com/trozmarynowski/Projekt-Zaliczeniony-Merito/refs/heads/main/BreedLabels.csv'
url_color = 'https://raw.githubusercontent.com/trozmarynowski/Projekt-Zaliczeniony-Merito/refs/heads/main/ColorLabels.csv'

df = pd.read_csv(url_train)
breed_labels = pd.read_csv(url_breed)
color_labels = pd.read_csv(url_color)

print(f"Rozmiar zbioru głównego: {df.shape}")\

df.head()
```

```python
# --- PRZYGOTOWANIE DANYCH---
# Zamieniam numery ras i kolorów na słowa, żeby wykresy były czytelne.


# 1. Wczytanie (korzystamy z linków, które zdefiniowałeś wyżej)
df = pd.read_csv(url_train)
breed_labels = pd.read_csv(url_breed)
color_labels = pd.read_csv(url_color)

# 2. Łączenie (Rasy i Kolory)
df = pd.merge(df, breed_labels, left_on='Breed1', right_on='BreedID', how='left')
df.rename(columns={'BreedName': 'Breed1_Name'}, inplace=True)

df = pd.merge(df, color_labels, left_on='Color1', right_on='ColorID', how='left')
df.rename(columns={'ColorName': 'Color1_Name'}, inplace=True)

# 3. Sprzątanie (usuwam tylko zbędne ID, 'Type' zostaje!)
df.drop(['BreedID', 'ColorID'], axis=1, inplace=True)


print(df.columns.tolist())
```

```python
# --- EKSPLORACYJNA ANALIZA DANYCH (EDA) ---


# Podczas tej analizy chcę sprawdzic jak szybko zwierzęta znajdują nowy dom.
# Legenda:
# 0 - W tym samym dniu
# 1 - 1-7 dni
# 2 - 8-30 dni
# 3 - 31-90 dni
# 4 - Brak adopcji po 100 dniach

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='AdoptionSpeed', data=df, palette='viridis')

plt.title('Rozkład szybkości adopcji (Zmienna celu)', fontsize=15)
plt.xlabel('Kategoria szybkości (0=Dziś ... 4=Powyżej 100 dni)', fontsize=12)
plt.ylabel('Liczba zwierząt', fontsize=12)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# TYP ZWIERZĘCIA (PSY vs KOTY) ---
# Sprawdzam, czy gatunek ma wpływ na szybkość adopcji.
# 1 = Pies, 2 = Kot.

# Tworzę kolumnę z nazwami, żeby na wykresie było "Pies/Kot" a nie "1/2"
df['Typ_Nazwa'] = df['Type_x'].map({1: 'Pies', 2: 'Kot'})

plt.figure(figsize=(10, 6))

# Rysuję wykres.
# x = Szybkość Adopcji
# hue = Podział na kolory wg Typu (Pies/Kot)
sns.countplot(x='AdoptionSpeed', hue='Typ_Nazwa', data=df, palette='pastel')

plt.title('Szybkość adopcji: Psy vs Koty', fontsize=15)
plt.xlabel('Szybkość Adopcji (0=Szybko ... 4=Wolno)', fontsize=12)
plt.ylabel('Liczba zwierząt', fontsize=12)
plt.legend(title='Gatunek')

plt.show()
```
